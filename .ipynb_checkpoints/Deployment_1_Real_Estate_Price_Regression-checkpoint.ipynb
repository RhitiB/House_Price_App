# Real Estate Price Prediction- Regression

The real estate valuation is a regression problem. The market historical data set of real estate valuation are collected from Sindian Dist., New Taipei City, Taiwan.

**Variable** \
The inputs are as follows
- X1=the transaction date (for example, 2013.250=2013 March, 2013.500=2013 June, etc.)
- X2=the house age (unit: year)
- X3=the distance to the nearest MRT station (unit: meter)
- X4=the number of convenience stores in the living circle on foot (integer)
- X5=the geographic coordinate, latitude. (unit: degree)
- X6=the geographic coordinate, longitude. (unit: degree)

The output is as follow
Y= house price of unit area (10000 New Taiwan Dollar/Ping, where Ping is a local unit, 1 Ping = 3.3 meter squared). Here Y is the target variable

```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt, numpy as np
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
real_estate_valuation = fetch_ucirepo(id=477) 
  
# data (as pandas dataframes) 
X = real_estate_valuation.data.features 
y = real_estate_valuation.data.targets 
df=pd.concat([X,y], axis=1)
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe().T
```

```python
columns=["X1", "X2", "X3", "X4", "X5", "X6", "Y"]
df.columns=columns
df.head()
```

```python
# Dropping the date column as it is not related to our data
df.drop("X1", axis=1, inplace=True)
```

```python
sns.pairplot(data=df)
plt.show()
```

We can observe a positive relation between X5 and X6 and the target variable Y. A negative relation is seen between x6 and X3. 

```python
corr=df.corr()
print(corr)
sns.heatmap(corr, cmap="Reds", annot=True)
```

```python
X_values = df.drop(columns=["Y"])   # keep as DataFrame
y_target = df["Y"]
#y_target=df["Y"]
#X_values=df.drop("Y", axis=1)
print(X_values.columns)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X_values, y_target, test_size=0.2, random_state=42)
print(X_train.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

## Models 
- Simple Linear Regression
- Linear Regression with Lasso and Polynomial fit
- Random Forest Regressor

```python
# Simple Linear Regression
from sklearn.linear_model import LinearRegression, Lasso
lr=LinearRegression()
lr.fit(X_train, y_train)
y_lr=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print("R square score", r2_score(y_lr, y_test))
print("MSE", mean_squared_error(y_lr, y_test))
```

```python
# Linear regression with Polynomial Fit
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3, include_bias=False)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
a_poly=lr.fit(X_train_poly, y_train)
y_lr_poly_train=lr.predict(X_train_poly)
y_lr_poly=lr.predict(X_test_poly)
print("R-Squared on test set", r2_score(y_test, y_lr_poly))
#print("R-Squared on train set", r2_score(y_lr_poly_train, y_train))
print("MSE on test set", mean_squared_error(y_test, y_lr_poly))
#print("MSE on train set", mean_squared_error(y_train, y_lr_poly_train))
```

```python
resid_train = y_train - y_lr_poly_train
resid_test  = y_test - y_lr_poly
# Plot
plt.figure(figsize=(5,3))
sns.scatterplot(x=y_lr_poly_train, y=resid_train, label="Train", alpha=0.7)
sns.scatterplot(x=y_lr_poly, y=resid_test, label="Test", alpha=0.7)
plt.axhline(0, ls="--", color="red")
plt.xlabel("Predicted values")
plt.ylabel("Residuals (Actual - Predicted)")
plt.title("Residual Plot")
plt.legend()
plt.show()
```

- Residuals are mostly around line 0 (red dashed line)
- No obvious scatter or pattern is observed so linearity looks okay
- Variance looks mostly stable

```python
# Linear Regression with L1 penalty
from sklearn.linear_model import LassoCV
alphas2 = np.array([1e-5, 5e-5, 0.0001, 0.0005])
LassoCV = LassoCV(alphas=alphas2, 
                  cv=4).fit(X_train_poly, y_train)
print(LassoCV.alpha_)
```

```python
lasso_best = Lasso(alpha=LassoCV.alpha_)
lasso_best.fit(X_train_poly, y_train)
y_lasso_pred = lasso_best.predict(X_test_poly)
#y_lasso_train_pred=lasso_best.predict(X_train_poly)
print("Lasso R²:", r2_score(y_test, y_lasso_pred))
print("Lasso MSE:", mean_squared_error(y_test, y_lasso_pred))
```

```python
#Doing a Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(
    n_estimators=600,   # number of trees
    max_depth=None,     # let it expand fully (can tune)
    random_state=42
)
rf.fit(X_train, y_train)
y_rf=rf.predict(X_test)
print("Random Forest R²:", r2_score(y_test, y_rf))
print("Random Forest MSE:", mean_squared_error(y_test, y_rf))
```

```python
f = plt.figure(figsize=(6,5))
ax = plt.axes()
ax.scatter(y_test, y_lr, label='Linear Regression',    s=18, marker='o',  alpha=0.7, zorder=2, color="yellow")
ax.scatter(y_test, y_lr_poly, label='Polynomial Regression', s=24, marker='x',  alpha=0.8, zorder=3, color="red")
ax.scatter(y_test, y_lasso_pred,label='Lasso Regression',      s=22, marker='D',  alpha=0.7, zorder=2, color="blue")
ax.scatter(y_test, y_rf,label='Random Forest Regression',      s=22, marker='*',  alpha=0.9, zorder=2, color="black")
ax.legend(frameon=True)
ax.set(xlabel='Actual Price', ylabel='Predicted Price', title='Regression Comparison')
plt.grid(True)
plt.show()
```

- The points are fairly in a straight line
- Polynomial and Lasso are very similar as points lie close to each other
- Linear looks the weakest model and Random Forest being the strongest as it also reduces overfitting of polynomial regression
- The R square and MSE of Random Forest are the best as compared to all other models
